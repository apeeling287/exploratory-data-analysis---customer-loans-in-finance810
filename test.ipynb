```python
# which of the correlated columns are most correlated with the rest of the dataset 
# correlation affects ML models which are used on the data, high correlation is detrimental to ML models 
# keep the column which is the least correlated
# above 0.7 is high correlation but does depend on the dataset e.g. scientific datasets need to be higher 
## two columns, drop one of them, calculate VIF against all other columns with each column, compare the VIFs - drop the bigger one - all in a for loop
# start with loan amount and funded amount, then run the heat map again and check for further correlations
# loan amount showed higher correlation than funded amount so dropped loan amount 
```

```python
import statsmodels.formula.api as smf
model1 = smf.ols("total_payment ~  funded_amount_inv + int_rate + annual_inc + dti + delinq_2yrs + inq_last_6mths + open_accounts + total_accounts + out_prncp + out_prncp_inv  + total_payment_inv + total_rec_prncp + total_rec_int + total_rec_late_fee + recoveries + collection_recovery_fee + last_payment_amount + collections_12_mths_ex_med + policy_code", modified_df).fit()
model1.summary()
```

```python
import statsmodels.formula.api as smf
model2 = smf.ols("total_payment_inv ~ funded_amount + funded_amount_inv + int_rate + annual_inc + dti + delinq_2yrs + inq_last_6mths + open_accounts + total_accounts + out_prncp + out_prncp_inv + total_payment + total_rec_prncp + total_rec_int + total_rec_late_fee + recoveries + collection_recovery_fee + last_payment_amount + collections_12_mths_ex_med + policy_code", modified_df).fit()
model2.summary()
```

```python
print(f"R^2 for model: \n total_payment: {model1.rsquared} \n total_payment_inv: {model2.rsquared}")
```

```python
## Code up a VIF function
def VIF(r2):
    return 1/(1-r2)

## Work out the VIF scores for each of the models
vif_funded_amount = VIF(model1.rsquared)
vif_instalment = VIF(model2.rsquared)


## print the VIF scores
print("VIF scores: \n total_payment: {}, \n total_payment_inv: {}".format(vif_funded_amount, vif_instalment))
```
